```python
sns.heatmap(df.corr(),annot=True,linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(20,18)
plt.show()
```

```python
sns.set()
plt.figure(figsize=(20,40))
count=1
for i in df.columns:
    plt.subplot(8,2,count)
    sns.distplot(df[i])
    count+=1
plt.show()
```
